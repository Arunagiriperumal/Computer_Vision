```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from keras.datasets import cifar10
from sklearn.metrics import classification_report
```

```python
dataset = cifar10.load_data()
```

###### Load , preprocess, 

```python
train_X = dataset[0][0]
train_Y = dataset[0][1]
test_X = dataset[1][0]
test_Y = dataset[1][1]
```

```python
train_X = train_X.reshape(train_X.shape[0], 32, 32, 3)
train_X = train_X / 255

test_X = test_X.reshape(test_X.shape[0], 32, 32, 3)
test_X = test_X / 255
```

# CNN
 
#Layer 1 ------------------------------Layer 2

#Number of filters = 32 -------------- Number of filters = 32

#kernel size 3 * 3  -------------------kernel size 3 * 3

#Activation function = Relu -----------Activation function = Relu

#Pooling : 2  ------------------------ Pooling : 2

#Padding : 0  ------------------------ Padding : 0
 
#Stride : 1 -------------------------- Stride : 1


Layer 3

Dense Layer : 64

Softmax output : 10

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Convolution2D, Dropout, Dense, Flatten, BatchNormalization, MaxPooling2D

#model
model_1 = Sequential()

model_1.add(BatchNormalization(input_shape=(32, 32, 3)))

#layer 1
model_1.add(Convolution2D(filters=32, kernel_size=3, activation="relu", input_shape=(32, 32, 3)))
model_1.add(MaxPooling2D(pool_size=2))
#layer 2
model_1.add(Convolution2D(filters=32, kernel_size=3, activation="relu", padding="same"))
model_1.add(MaxPooling2D(pool_size=2))
#Flatten
model_1.add(Flatten())
#FFN
model_1.add(Dense(units=64, activation="relu"))
#Classification Layer
model_1.add(Dense(units=64, activation="softmax"))
```

```python
from tensorflow.keras.optimizers import Adam
optimizer = Adam()
model_1.compile(optimizer=optimizer, loss="sparse_categorical_crossentropy", metrics=["accuracy"])
model_1.summary()
```

### Fitting a Model

```python
Model1 = model_1.fit(train_X, train_Y, epochs=15, validation_data=(test_X, test_Y), initial_epoch=0)
```

### CNN 2

#Layer 1 ------------------------------Layer 2 ------------------Layer 3

#Number of filters = 64 -------------- Number of filters = 32 ------------Number of filters = 16

#kernel size 3 * 3 -------------------kernel size 3 * 3 -------------------kernel size 3 * 3

#Activation function = Relu -----------Activation function = Relu -----------Activation function = Relu

#Pooling : 3 ------------------------ Pooling : 3 ------------------------ Pooling : 3

#Padding : 0 ------------------------ Padding : 0 ------------------------ Padding : 0

#Stride : 1 -------------------------- Stride : 1 -------------------------- Stride : 1

Dense layer : 64

Dense layer : 32

Dense layer : 16

Softmax output unit : 10

```python
model_2 = Sequential()

model_2.add(BatchNormalization(input_shape=(32, 32, 3)))

model_2.add(Convolution2D(filters=64, kernel_size=3, activation="relu", input_shape=(32,32,3)))
model_2.add(MaxPooling2D(pool_size=3))

model_2.add(Convolution2D(filters=32, kernel_size=3, padding="same", activation="relu"))
model_2.add(MaxPooling2D(pool_size=3))

model_2.add(Convolution2D(filters=16, kernel_size=3, activation="relu"))
#model_2.add(MaxPooling2D(pool_size=3))

model_2.add(Flatten())

model_2.add(Dense(units=64, activation="relu"))
model_2.add(Dense(units=32, activation="relu"))
model_2.add(Dense(units=16, activation="relu"))

model_2.add(Dense(units=10, activation="softmax"))
```

```python
adam = Adam()
model_2.compile(optimizer=adam, loss="sparse_categorical_crossentropy", metrics=["accuracy"])
model_2.summary()
```

```python
MODEL2 = model_2.fit(train_X, train_Y, epochs=5, validation_data=(test_X, test_Y), initial_epoch=0)
```

```python
EPOCHS = 5
loss_train = MODEL2.history["loss"]
loss_val = MODEL2.history["val_loss"]
epochs = range(1, EPOCHS + 1)
plt.plot(epochs, loss_train, "g", label="Training loss" )
plt.plot(epochs, loss_val, "b", label="Validation loss")
plt.title("Training and Validation loss")
plt.xlabel("EPOCHS")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

```python
EPOCHS = 5
Acc_train = MODEL2.history["accuracy"]
Acc_val = MODEL2.history["val_accuracy"]
epochs = range(1, EPOCHS + 1)
plt.plot(epochs, Acc_train, "g", label="Training Accuracy" )
plt.plot(epochs, Acc_val, "b", label="Validation Accuracy")
plt.title("Training and Validation Accuracy")
plt.xlabel("EPOCHS")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```
